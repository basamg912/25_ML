```python
from ultralytics import YOLO
model = YOLO("/content/drive/MyDrive/yolo_pretrained.pt")
```

```python
model.model
```

```python
for param in model.model.parameters():
  param.requires_grad = False
for param in model.model.model[-1].parameters():
  param.requires_grad = True
```

```python
model.train(
    data = "/content/drive/MyDrive/기계학습/pre_train/data.yaml",
    epochs = 10,
    batch = 10,
    imgsz = 640,
    device = 0
)
```

```python
model.save("fine_tune_yolo.pt")
```

```python
model_test = YOLO("/content/fine_tune_yolo.pt")
```

```python
outputs = model_test("/content/hoodie.jpeg")
```

```python
# 1. 탐지된 객체별 클래스명 출력
for result in outputs:
    boxes = result.boxes
    if boxes is not None:
        class_ids = boxes.cls.cpu().numpy().astype(int)  # 클래스 인덱스 배열
        names = model.names  # 클래스 id-이름 매핑 dict
        class_names = [names[cid] for cid in class_ids]
        print("Detected classes:", class_names)

# 2. 간단 요약 정보 (탐지 객체 수 등)
print(f"Total detections: {len(boxes)}")
```
